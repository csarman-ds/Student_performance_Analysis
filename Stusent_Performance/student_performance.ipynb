```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
#load data set
df=pd.read_csv("StudentsPerformance.csv")
#explore dataset
df.head()
df.info()
df.describe()
df.isnull().sum()
#label encoder
encoder=LabelEncoder()
df["gender"]=encoder.fit_transform(df["gender"])
df["race/ethnicity"]=encoder.fit_transform(df["race/ethnicity"])
df["parental level of education"]=encoder.fit_transform(df["parental level of education"])
df["lunch"]=encoder.fit_transform(df["lunch"])
df["test preparation course"]=encoder.fit_transform(df["test preparation course"])
#data visualization
#LINE PLOT
plt.figure(figsize=(8,5))
sns.lineplot(x="reading score",y="math score",data=df)
plt.title("reading score vs math score")
plt.show()
#SCATTER PLOT
plt.figure(figsize=(6,4))
sns.scatterplot(x="reading score",y="math score",data=df)
plt.title("reading score vs math score")
plt.show()
#select x and y
x=df.drop("math score", axis=1)
y=df["math score"]
#split data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#create model
model=LinearRegression()
#train model
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
#evaluate
print("mse:",mean_squared_error(y_test,y_pred))
print("mae:",mean_absolute_error(y_test,y_pred))
print("R2_score:",r2_score(y_test,y_pred))
joblib.dump(model,"Student_Performance.pkl")
print("model saved sucessfully!")
new_data=pd.DataFrame({
    "gender":[1],
    "race/ethnicity":[2],
    "parental level of education":[3],
    "lunch":[1],
    "test preparation course":[1],
    "reading score":[80],
    "writing score":[85]})
prediction=model.predict(new_data)
print("predicted math score:",prediction)
```
